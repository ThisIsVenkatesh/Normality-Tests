# Q-Q Plot using scipy.stats.probplot()

```python
#Q-Q plot
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats

# N(0,1)
#std_normal = np.random.normal(loc = 0, scale = 1, size=1000)

# 0 to 100th percentiles of std-normal
#for i in range(0,101):
#    print(i, np.percentile(std_normal,i))
```

```python
# generate 5 samples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size=5) 
#try size=1000
print(min(measurements),max(measurements))

A=stats.probplot(measurements, dist="norm", plot=plt)
print(type(A),len(A)) #returns a tuple of size 2
print('-'*20)
#---------------------------------------------------------------------------------------------
print(len(A[0])) #the first element in the tuple contains two ndarrays, in that the first one is quantiles of 
# generated values(number of generated values=size of input array) from specified theoretical distribution,
# the second one is the sorted values of the input array 

print(A[0][0]) #First it generates the theoretical values with the same size that of the input array, then 
#then quantiles are calculated(# of quantiles = size of input array), the quantiles are
#calculated by Filliben’s estimate, these quantiles are called (osm - ordered statistic medians) 

print(A[0][1]) #the sorted version of input array, each value become an quantile (osr - ordered responses)
print('-'*20)

# the returned plot is actually equivalent to the scatterplot(A[0][0],A[0][1])
# the theoretical distribution quantiles are taken horizontally.  

#---------------------------------------------------------------------------------------------
print(f'{len(A[1])} elements in the 2nd element in tuple')
#Only returns if 'fit=True'
print(A[1][0]) # interceptor obtained by principle of least squares
print(A[1][1]) # slope obtained by principle of least squares
print(A[1][2]) # r-square value

plt.show()
```

```python
# generate 10 samples from N(20,5)
measurements = np.random.uniform(low=-1, high=1, size=10) 
#try size=1000

A=stats.probplot(measurements, dist="norm", plot=plt)
print(A[1][2]) #printing r-square value
plt.show()
```

# Q-Q Plot using statsmodels.graphics.gofplots.qqplot()

```python
from statsmodels.graphics.gofplots import qqplot
qqplot(measurements, line='s')
plt.show() # both methods(scipy and statsmodels methods) are similar
```
