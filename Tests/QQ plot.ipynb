# Q-Q Plot using scipy.stats.probplot()

```python
# Q-Q plot
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

```python
# Generates 5 samples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size=5) 
#try size=1000
print(min(measurements),max(measurements))

A=stats.probplot(measurements, dist="norm", plot=plt)
print(type(A),len(A)) #returns a tuple of size 2
print('-'*20)
#---------------------------------------------------------------------------------------------
print(len(A[0])) #the first object in the tuple contains two nd-arrays, the first one contains
# generated quantiles from referece distribution(number of generated values=size of input array)
# the second one is the sorted values of the input array

print(A[0][0]) #First the algorithm generates n probability values by Filliben’s estimate 
# form that n(the size of input array) probabilities we will get n quantiles from desired distribution
# with the use of the quantile function of that distribution.
# please refer my [answer] on stackoverflow for more information.
# https://stackoverflow.com/questions/58369546/weird-behavior-from-scipy-stats-probplot/58994608#58994608

print(A[0][1]) #the sorted values of input array, each value become an quantile(osr - ordered responses)
print('-'*20)

# the returned plot is actually equivalent to the scatterplot(A[0][0],A[0][1])
# the theoretical distribution quantiles are taken horizontally.  

#---------------------------------------------------------------------------------------------
print(f'{len(A[1])} elements in the 2nd element in tuple')
#Only returns if 'fit=True' - The results obtained by fitting a st.line
print(A[1][0]) # interceptor obtained by principle of least squares
print(A[1][1]) # slope of fitted line obtained by principle of least squares
print(A[1][2]) # R-square value

plt.show()
```

```python
# Generate 10 samples from N(20,5)
measurements = np.random.uniform(low=-1, high=1, size=10) 
#try size=1000

A=stats.probplot(measurements, dist="norm", plot=plt)
print(A[1][2]) #printing R-square value
plt.show()
```

# Q-Q Plot using statsmodels.graphics.gofplots.qqplot()

```python
from statsmodels.graphics.gofplots import qqplot
qqplot(measurements, line='s')
plt.show() # both methods(scipy and statsmodels methods) gives similar results.
# Note: The normal probability plot is a special case of the Q–Q probability plot for a normal distribution.
```
